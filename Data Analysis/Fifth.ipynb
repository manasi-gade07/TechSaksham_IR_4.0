```python
f = df['Fuel_Type'].value_counts()
plt.title('Fuel Distribution')
plt.xlabel('car_count')
colors = ['yellow','blue','red']
f.plot(kind='bar',color=colors)

plt.plot()
plt.grid(linestyle='dotted')
```

```python
t=df['Transmission'].value_counts()
colors = ['yellow','blue']
plt.title('Transmission analysis')
plt.ylabel('Transmission count')

t.plot(kind='bar',color=colors)
plt.show()
```

```python
t=df['Transmission'].value_counts()
colors = ['yellow','red']
plt.title('Transmission analysis')
plt.ylabel('Transmission count')
plt.pie(t,colors=colors)
plt.show()
```
