```python
## Copy and View

# Copy does not affect original array 

a = np.array([1,2,3,4,5])
c = a.copy()
a[0]=56
c[4] = 67
print('original',a)
print('copy',c)


# View affect original array

a = np.array([1,2,3,4,5])
c = a.view()
c[4] = 67
a[0]=56
print('original',a)
print('view',c)
```

```python
## Student Data and Marks

marks = np.array([40,50,45,60,88,90,100,0])
students = np.array([10,20,30,40,50,60,70,80])
plt.grid()
plt.plot(marks,students,marker='o',markerfacecolor='red',markersize=6,markeredgecolor='red')

plt.show()
```

```python
## Bar Chart

marks = np.array([40,50,45,60,88,90,100,0])
student = np.array([10,20,30,40,50,60,70,80])
plt.grid()
plt.bar(marks,student)

plt.show()
```

```python
x = np.array([1,2])
y = np.array([5,8])
colors = ['red','yellow']
plt.bar(x,y,color=colors)

plt.show()
```

```python
x = np.array([1,2])
y = np.array([5,8])
colors = ['yellow','blue']
plt.barh(x,y,color=colors)
plt.xlabel('marks')
plt.ylabel('student data')
plt.show()
```

```python
## Scattered Plot

x = np.array([1,3,5,4,6])
y = np.array([1,3,5,7,8])
colors = ['red','green','blue','purple','yellow']
plt.grid(linewidth=0.5,linestyle='dotted')
plt.scatter(x,y,color=colors,linewidth=1)
plt.show()
```

```python
## Pie Chart

x = np.array([25,25,25,25])
colors = ['red','blue','yellow','green']
label = ['red','blue','yellow','green']
plt.pie(x,colors=colors,labels=label)
plt.legend()
plt.show()
```
