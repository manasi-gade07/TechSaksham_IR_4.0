```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.array([1,2,3])
y = np.array([4,2,3])
plt.plot(x,y)
plt.show()
```

```python
x = np.array([1,2,3])
y = np.array([4,2,3])
df = pd.DataFrame({
'x': x,
'y' : y
})
sns.lineplot(x='x',y='y',data=df)
plt.show()
```

```python
plt.plot(df['x'],df['y'])
```

```python
df = sns.load_dataset('tips')
```

```python
plt.figure(figsize=(4,4))
sns.barplot(x='time',y='total_bill',data=df,palette='coolwarm',hue='sex')
plt.show()
```

```python
plt.figure(figsize=(4,4))
sns.countplot(x='time',data=df,color='red')
plt.show()
```

```python
plt.figure(figsize=(4,4))
sns.lineplot(x='time',y='total_bill',data=df,hue='sex',style='sex',legend=False)
plt.show()
```

```python
plt.figure(figsize=(4,50))
sns.barplot(x='time',y='total_bill',data=df,palette='coolwarm',hue='sex',orient='h')
plt.show()
```

```python


sns.histplot(df['total_bill'],kde=True)
plt.grid()
plt.show()
```

```python
sns.scatterplot(x='total_bill',y='tip',data=df,hue='sex',style='sex')
plt.show()
```

```python
sns.boxplot(x='time',y='tip',data=df,hue='sex')
plt.show()
```

```python
sns.violinplot(x='time',y='tip',data=df,hue='sex',inner=None,split=True)
plt.show()
```

```python
a = np.array([[1,2,3],[1,2,3],[5,2,3],[1,5,3]])
a2 = {'fontsize':10 }
sns.heatmap(a,annot=a,annot_kws=a2,linewidth=4,linecolor='yellow')
plt.show()
```

```python
a = sns.pairplot(df,vars=['total_bill','tip'])
plt.show()
```

```python
a = sns.stripplot(x='total_bill',y='tip',data=df)
plt.show()
```

```python
a = sns.catplot(x='total_bill',y='tip',data=df)
plt.show()
```
