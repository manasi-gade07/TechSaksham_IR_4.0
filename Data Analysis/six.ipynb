
## Ship Mode Analysis based on Sales and Profit


```python
s = df['Ship Mode'].value_counts()
```

```python
print(s)
```

```python
plt.figure(figsize=(4,4))
c=['Green','yellow','blue','red']
s.plot(kind='bar',color=c)
plt.title('Ship Mode Analysis',loc='right')
plt.ylabel('count')
plt.show()
```

```python
a = df.groupby('Ship Mode')['Sales'].sum()      ## Finding more than one attribute insights
print(a)
```

```python

a.plot(kind='bar')
plt.show()
```

```python

a.plot(kind='pie',autopct='%1.1f%%')
plt.legend()
plt.title('Ship Mode Analysis based on sales',loc='right')

plt.show()
```


## Segment Analysis based on Sales,Profit and Discount


```python
a = df.groupby('Ship Mode')['Profit'].sum()      
print(a)
```

```python

c=['Green','yellow','blue','red']
a.plot(kind='bar',color=c)
plt.title('Ship Mode Analysis based on profit',loc='right')
plt.ylabel('count')
plt.show()
```

```python
a = df.groupby('Segment')['Sales'].sum()   
```

```python
print(a)
```

```python
c=['Green','yellow','blue','red']
a.plot(kind='bar',color=c)
plt.title('Segment Analysis based on sales',loc='right')
plt.ylabel('count')
plt.show()
```

```python
a = df.groupby('Segment')['Profit'].sum()   
print(a)
```

```python
c=['Yellow','red','blue']
a.plot(kind='pie',colors=c,autopct='%1.1f%%')
plt.legend()
plt.title('Segment Analysis based on profit',loc='right')
plt.ylabel('Sales count')
plt.show()
```

```python
a = df.groupby('Segment')['Discount'].sum()   
print(a)
```

```python
c=['Yellow','red','blue']
a.plot(kind='bar',color=c)
plt.title('Segment Analysis based on profit',loc='right')
plt.ylabel('Sales count')
plt.show()
```

## Region Analysis based on Sales and Profit

```python
c=['Yellow','red','blue']
a.plot(kind='bar',color=c)
plt.title('Region Analysis based on sales',loc='right')
plt.ylabel('Sales count')
plt.show()
```

```python
a = df.groupby('Region')['Profit'].sum()  
print(a)
```
